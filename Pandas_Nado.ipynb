# DataFrame

```python
# DataFrame
# Dictionary를 통해 자료 구조 생성
# 2차원 데이터의 표현 


data = {
    '이름': ['유재석', '강호동','송혜교','정우성','강동원'],
    '학교': ['A', 'B', 'C', 'D', 'E'],
    '국어': [100, 90, 90, 90, 80],
    '수학': [70, 80, 100, 90, 100],
    '영어': [60, 80, 90, 100, 97]
    }



```
