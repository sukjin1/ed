#원그래프 기본

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
plt.pie(values, labels = labels, autopct="%.2f%%", startangle=0, counterclock=False)
plt.show()

```

```python
explode=[0.2, 0.1, 0, 0, 0, 0]
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title = '언어별 선호도')
plt.show()
```

#원 그래프 심화

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y']
colors = ['#ffadad','#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

explode = [0.05] * 6
plt.pie(values, labels = labels, autopct="%.2f%%", startangle=0, colors = colors, counterclock=False, explode=explode)
plt.show()
```

```python
wedgeprops = {'width':0.3, 'edgecolor':'black', 'linewidth': 1}

plt.pie(values, labels = labels, autopct="%.2f%%", startangle=0, colors = colors, 
        wedgeprops = wedgeprops, counterclock=False, explode=explode)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct>= 10 else ''

# plt.pie(values, labels = labels, autopct="%.2f%%", startangle=0, colors = colors, 
        # wedgeprops = wedgeprops, counterclock=False, explode=explode)
plt.pie(values, labels = labels, pctdistance=0.5, autopct=custom_autopct, startangle=0, colors = colors, 
        wedgeprops = wedgeprops, counterclock=False, explode=explode)
plt.show()
```

#평점이 9점 이상인 영화와 9점 미만인 영화를 구분하여 각각의 비율을 원그래프로 출력

범례: 그래프 오른쪽에 표시

```python
filt = df['평점'] >= 9

print(len(df[filt]))
print(len(df[~filt]))

labels = ['9점이상', '9점미만']
plt.pie([len(df[filt]), len(df[~filt])], 
        autopct = "%.1f%%" )
plt.legend(loc=(1, 0.2), labels=labels)

```
