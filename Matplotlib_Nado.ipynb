#Multiple Data

##COVOD-19 백신 종류별 접종 인구

```python
days = [1,2,3] # 1일, 2일, 3일
az = [2,4,8] # 단위:만명
pfizer = [5,1,3] 
moderna = [1,2,5]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker = 'o', ls='--')
plt.plot(days, moderna, label='moderna', marker = 's', ls='-')

plt.legend(ncol=3)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker = 'o', ls='--')
plt.plot(days, moderna, label='moderna', marker = 's', ls='-')
plt.legend()
```

#막대그래프

```python
labels = ['강백호', '서태웅', '정대만']
values = [188, 190, 180]
colors = ['r','g', 'b']
# plt.bar(labels, values, color='r', )
plt.bar(labels, values, color=colors, alpha=0.5 )
```

```python
plt.bar(labels, values, color=colors, alpha=0.5 )
plt.ylim(160, 200)
```

```python
plt.bar(labels, values, color=colors, alpha=0.5, width=0.5 )
plt.ylim(160, 200)
```

```python
ticks = ['1번학생', '2번학생', '3번학생']
plt.bar(labels, values, color=colors, alpha=0.5, width=0.5 )
plt.xticks(labels, ticks, rotation = 90) # x축의 데이터 각도
plt.yticks(rotation = 50) # x축의 데이터 각도
```
