```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def gradient_descent(x, y, learning_rate, num_iterations):
    num_samples = x.shape[0]
    num_features = x.shape[1]
    theta = np.zeros(num_features + 1)  # Initializing weights with bias term

    # Adding a column of 1s for the bias term in the input data
    x_with_bias = np.concatenate((np.ones((num_samples, 1)), x), axis=1)

    for iteration in range(num_iterations):
        z = np.dot(x_with_bias, theta)
        h = sigmoid(z)
        gradient = 1 / num_samples * np.dot(x_with_bias.T, (h - y))
        theta -= learning_rate * gradient
        print(theta)
        # Plotting the data points and decision boundary
        plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.RdYlBu)
        x_boundary = np.array([np.min(x[:, 0]), np.max(x[:, 0])])
        y_boundary = -(theta[0] + theta[1] * x_boundary) / theta[2]
        plt.plot(x_boundary, y_boundary, 'k-', alpha=0.01)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.title('Logistic Regression')
#        plt.show()

    return theta

# Data preparation
x = np.array([[2, 3], [4, 1],[2, 2], [3, 1], [6, 7], [8, 5], [7, 7], [9, 5]])  # Input variable (2-dimensional features)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])  # Output variable (binary labels)

# Hyperparameter settings
learning_rate = 0.3
num_iterations = 100

# Gradient descent execution
theta = gradient_descent(x, y, learning_rate, num_iterations)

print(theta)
# Plotting the data points and decision boundary
plt.show()
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.RdYlBu)
x_boundary = np.array([np.min(x[:, 0]), np.max(x[:, 0])])
y_boundary = -(theta[0] + theta[1] * x_boundary) / theta[2]
plt.plot(x_boundary, y_boundary, 'k-', alpha=0.1)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression')
plt.show()
```
