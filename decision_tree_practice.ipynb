```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:] #꽃의 길이와 넓이
y = iris.target # 종류를 라벨로

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)


tree_clf = DecisionTreeClassifier(max_depth=2) #트리의 깊이를
tree_clf.fit(X_train, y_train)

y_pred = tree_clf.predict(X_test)

# Evaluating the Algorithm
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, :] #꽃받침/꽃의 길이와 넓이
y = iris.target # 종류를 라벨로

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)


tree_clf = DecisionTreeClassifier(criterion = "entropy", max_depth=3) #
tree_clf.fit(X_train, y_train)

y_pred = tree_clf.predict(X_test)

# Evaluating the Algorithm
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:,0:2] #꽃받침 데이터
y = iris.target # 종류를 라벨로

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)


tree_clf = DecisionTreeClassifier(criterion = "entropy", max_depth=3) #
tree_clf.fit(X_train, y_train)

y_pred = tree_clf.predict(X_test)

# Evaluating the Algorithm
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

tree_clf.feature_importances_
from sklearn import tree
import graphviz
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 8))
_ = tree.plot_tree(tree_clf,
                  feature_names=['x1', 'x2'],
                  class_names=['0', '1', '2'],
                  filled=True)
```
